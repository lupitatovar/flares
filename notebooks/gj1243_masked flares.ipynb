```python
#creates array same shape as kepler time (or whatever data you give)
f_mask = np.zeros_like(kepler_time, dtype=bool)
```

```python
#mask array
for i in range(len(t_starts)):
    temp = (kepler_time >= t_starts[i]) & (kepler_time <= t_stops[i])
    f_mask = f_mask + temp
print(f_mask)
```

```python
#plot with no flares but over flattend

no_flare = np.where((f_mask == False))[0]
plt.scatter(kepler_time[no_flare],kepler_flux[no_flare],marker='o',alpha=0.5)
```

```python
# run celerite on this data (has masked flares!!)
#plot shows one day of (non-flat) flare-less data
plt.scatter(gj1243lc.time[no_flare],gj1243lc.flux[no_flare]/np.nanmedian(gj1243lc.flux))
plt.xlim(539.4,540.4)
plt.ylim(1.17,1.22)
```
